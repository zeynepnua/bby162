```
from random import randint



print("Oyunda '_'  karakteri tahmin edilmesi gereken harf, '|' karakteri ise kelimeler arasındaki boşluğu ifade eder")

print("Oyun Başlıyor...")



kalanCan = 3

tahminEdilenHarfler = []

kelimeler = ["Türkiye", "Fransa", "Çin", "Wuhan", "Amerika", "Fahrettin Koca", "Vaka Sayısı",

             "Karantina", "Okul", "Özlem", "Can Sıkıntısı", "Tom Hanks", "Fatih Terim"]

secilenKelime = kelimeler[randint(0, len(kelimeler) - 1)]

gosterilecekKelime = ""



for i in range(len(secilenKelime)):

    if (secilenKelime[i] == " "):

        gosterilecekKelime += "|"

    else:

        gosterilecekKelime += "_"



oyunBittiMi = False



while not oyunBittiMi:

    print("Kalan Can : ", kalanCan)

    print("Tahmin Edilecek Kelime:")

    for i in range(len(secilenKelime)):

        print(gosterilecekKelime[i], end=' ')



    tahmin = input("\nTahminizi giriniz: ")



    if tahmin in tahminEdilenHarfler:

        tahmin = input("Bu harfi daha önce denemiştiniz :) Yeni Tahminizi giriniz: ")

    else :

        tahminEdilenHarfler.append(tahmin)



    if kalanCan == int(0):

        print("Oyunu Kaybettin :( Aranan Kelime " + secilenKelime + " idi")

        oyunBittiMi = True

        break



    if secilenKelime.lower().find(tahmin.lower()) == int(-1) :

        print("YANLIŞ TAHMİN!!!!...")

        kalanCan -= 1

    else :

        for i in range(len(secilenKelime)):

            if (tahmin.lower() == secilenKelime[i].lower()):

                gosterilecekKelime = gosterilecekKelime[:i] + secilenKelime[i] + gosterilecekKelime[i + 1:]



    if gosterilecekKelime.find("_") == int(-1):

        for i in range(len(secilenKelime)):

            if(gosterilecekKelime[i] == "|"):

                print(" ", end = ' ')

            else :

                print(gosterilecekKelime[i], end=' ')

        print("\nOYUNU KAZANDIN BRAVO! Aranan Kelime " + secilenKelime + " idi :)")

        oyunBittiMi = True





# print(secilenKelime)
```
