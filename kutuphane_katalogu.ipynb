```
import sys

from google.colab import drive

drive.mount("/gdrive",force_remount=True)





veritabani = "/gdrive/My Drive/veritabani.txt"



#ornek data

ornek="Suç ve Ceza\tFyodor Dostoyevski\ttest\t1990\nSefiller\tVictor Hugo\ttest\t1995\n"



with open(veritabani, 'w') as file:

    file.write( ornek)



while True:

    islem = int(input("Lutfen bir islem seciniz:\n"

                  "Katalogtaki eserleri listelemek icin: 1\n" 

                  "Kataloga kitap eklemek icin: 2\n"

                  "Katalogtaki kitabi düzenlemek icin: 3\n"

                  "Katalogtaki kitabi silmek icin: 4\n"

                  "Katalogta kitap aramak icin: 5\n"

                  "Cikis yapmak icin: 6\n"))

    if (islem == 1):

        print("\n\n")

        file = open(veritabani, "r")

        print(file.read())

        print("\n\n")

    elif(islem == 2):

        print("\n\n")

        bookName= input("Girmek istediginiz kitabin ismi: ")

        writer = input("Girmek istediginiz kitabin yazari: ")

        publisher = input("Girmek istediginiz kitabin yayinevi: ")

        year = input("Girmek istediginiz kitabin yayim yili: ")

        with open(veritabani, 'a+') as file:

            file.write( str(bookName) + "\t" + str(writer) + "\t" + str(publisher) + "\t" + str(year) + "\n")

        print("Kitap Kataloga Eklendi\n")

    elif(islem==3):

        print("\n\n")

        file = open(veritabani, "r")

        txt=file.read()

        books= txt.split("\n")

        i=0;

        for x in books:

            if(x != ""):

                i=i+1

                print(str(i)+")" + x)

        no=int(input("\nDuzeleme yapmak istediginiz eserin numarasini giriniz: " ))

        bookName= input("Kitabin ismi: ")

        writer = input("Kitabin yazari: ")

        publisher = input("Kitabin yayinevi: ")

        year = input("Kitabin yayim yili: ")

        books[no-1]=str(bookName) + "\t" + str(writer) + "\t" + str(publisher) + "\t" + str(year)

        with open(veritabani, 'w') as file:

            for x in books:

                if(x != ""):

                    file.write( x + "\n")

            print("Eser duzenlendi\n")

    elif(islem==4):

        print("\n\n")

        file = open(veritabani, "r")

        txt=file.read()

        books= txt.split("\n")

        i=0;

        for x in books:

            if(x != ""):

                i=i+1

                print(str(i)+")" + x)

        no=int(input("\nSilmek istediginiz eserin numarasini giriniz: " ))

        del books[no-1]

        with open(veritabani, 'w') as file:

            for x in books:

                if(x != ""):

                    file.write( x + "\n")

            print("Eser silindi\n")

    elif(islem==5):

        print("\n\n")

        search= input("Arama islemi : ")

        file = open(veritabani, "r")

        txt=file.read()

        books= txt.split("\n")

        i=0;

        for x in books:

            if(x != "" and x.find(search)!=-1 ):

                i=i+1

                print(str(i)+")" + x)

        print("\n\n")

    elif(islem == 6):

        print("Cikis Yapiliyor")

        break;

    else:

        print("Yanlis secim yaptiniz. Lutfen tekrar seciniz.")


```
